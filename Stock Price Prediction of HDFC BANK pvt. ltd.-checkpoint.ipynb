```python
from statsmodels.tsa.stattools import adfuller
def adf_test(testdata):
    out=adfuller(testdata)
    labels=['ADF Statistic','P-Value','Lags','Observations']
    for value,label in zip(out,labels):
        print(label,':',value)
    if out[1]<0.05:
        print('Reject Null Hypothesis as data has no unit root and is stationary')
    else:
        print('Accept Null Hypothesis as data has a unit root and is non-stationary')
    
```

```python
adf_test(data1['VWAP'])
```

```python
data1['Date']=pd.to_datetime(data1['Date'])
data2=data1.copy()
```

```python
data2['Month']=data2["Date"].dt.month
data2['Day']=data2["Date"].dt.day
data2['Year']=data2["Date"].dt.year
data2['DayofWeek']=data2["Date"].dt.dayofweek
```

```python
data2.head()
```

```python
inputs= ["High", "Low", "Volume", "Turnover","Close"]
w1,w2,w3=3,7,30
#applying rolling techniques
data2_3days = data2[inputs].rolling(window=w1, min_periods=0)
data2_7days = data2[inputs].rolling(window=w2, min_periods=0)
data2_30days = data2[inputs].rolling(window=w3, min_periods=0)
# finding the means of the rolled data
data2_avg_of_3days = data2_3days.mean().shift(1).reset_index()
data2_avg_of_7days = data2_7days.mean().shift(1).reset_index()
data2_avg_of_30days = data2_30days.mean().shift(1).reset_index()
#finding the standard deviation of the rolled data
data2_std_of_3days = data2_3days.std().shift(1).reset_index()
data2_std_of_7days = data2_7days.std().shift(1).reset_index()
data2_std_of_30days = data2_30days.std().shift(1).reset_index()
```

```python
for features in inputs:
    # arrangement of features my their means and creating separate features for each
    data2[f"{features}_mean_lag{w1}days"] = data2_avg_of_3days[features]
    data2[f"{features}_mean_lag{w2}days"] = data2_avg_of_7days[features]
    data2[f"{features}_mean_lag{w3}days"] = data2_avg_of_30days[features]
    # arrangement of features my their standard deviation and creating separate features for each
    data2[f"{features}_std_lag{w1}days"] = data2_std_of_3days[features]
    data2[f"{features}_std_lag{w2}days"] = data2_std_of_7days[features]
    data2[f"{features}_std_lag{w3}days"] = data2_std_of_30days[features]
```

```python
train_data=data2[data2['Date']<'2018']
test_data=data2[data2['Date']>='2018']
```

```python
train_data
```

```python
data2.columns
```

```python
exogenous_features=['Month',
       'Day', 'Year', 'DayofWeek', 'High_mean_lag3days', 'High_mean_lag7days',
       'High_mean_lag30days', 'High_std_lag3days', 'High_std_lag7days',
       'High_std_lag30days', 'Low_mean_lag3days', 'Low_mean_lag7days',
       'Low_mean_lag30days', 'Low_std_lag3days', 'Low_std_lag7days',
       'Low_std_lag30days', 'Volume_mean_lag3days', 'Volume_mean_lag7days',
       'Volume_mean_lag30days', 'Volume_std_lag3days', 'Volume_std_lag7days',
       'Volume_std_lag30days', 'Turnover_mean_lag3days',
       'Turnover_mean_lag7days', 'Turnover_mean_lag30days',
       'Turnover_std_lag3days', 'Turnover_std_lag7days',
       'Turnover_std_lag30days', 'Close_mean_lag3days', 'Close_mean_lag7days',
       'Close_mean_lag30days', 'Close_std_lag3days', 'Close_std_lag7days',
       'Close_std_lag30days']
```

```python
from pmdarima import auto_arima
auto_ar_model=auto_arima(train_data['VWAP'], exogenous=train_data[exogenous_features],
                         trace=True, error_action="ignore", suppress_warnings=True)
auto_ar_model.fit(train_data['VWAP'], exogenous=train_data[exogenous_features])
test_data['Forecast']=auto_ar_model.predict(n_periods=len(test_data),exogenous=test_data[exogenous_features])
```

```python
auto_ar_model.summary()
```

```python
test_data[["VWAP", "Forecast"]].plot(figsize=(15,9))
```

```python
errors=test_data['VWAP']-test_data['Forecast']
```

```python
adf_test(errors)
```

```python
errors.plot(figsize=(15,9))
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(test_data['VWAP'], test_data['Forecast'])))
print("Mean Absolute Error:", mean_absolute_error(test_data['VWAP'], test_data['Forecast']))
```
